## Setup



```python
from datasets import load_dataset
from icecream import ic

from util import *

```

## kilt_tasks



```python
dnm = 'kilt_tasks'
dset = load_dataset(dnm, name='nq')
for dnm, dset_ in dset.items():
    ic(dnm, dset_)
    for i in range(2):
        d = dset_[i]
        ic(d['input'], [elm['answer'] for elm in d['output']])

```

## wiki_qa



```python
dnm = 'wiki_qa'
dset = load_dataset(dnm)
for dnm, dset_ in dset.items():
    ic(dnm, dset_, dset_[:4])
    for i in range(4):
        d = dset_[i]
        ic(d['question'], d['answer'])

```
